```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# prevents warnings about future implementations from appearing
import warnings
warnings.filterwarnings('ignore', 'use_inf_as_na')

iris = sns.load_dataset('iris')
print(iris)
```

```python
plt.boxplot(iris['sepal_length'], showmeans=True, meanline=True)
plt.show()
```

```python
iris.boxplot('sepal_length', showmeans=True, meanline=True, by='species')
plt.show()
```

```python
sns.boxplot(data=iris, x='species', y='sepal_length', hue='species', dodge=False, width=.3)
plt.show()
```

```python
sns.boxenplot(data=iris, x='species', y='sepal_length', hue='species', dodge=False, width=.3)
```

```python
#like boxplots but curved which gives a better perspective of distribution of data
sns.violinplot(data=iris, x='species', y='sepal_length', hue='species', dodge=False)
```

```python
#generates dots to show distribution of data along the y axis similar to a violin plot but more granular and distinct,
#for example the green distribution is not actually that dense in the upper quartile
#does not always scale well
sns.swarmplot(data=iris, x='species', y='sepal_length', hue='species', dodge=False)
```

```python
#plots can be combined
sns.catplot(data=iris, x='species', y='sepal_length', kind='boxen', dodge=False)
sns.swarmplot(data=iris, x='species', y='sepal_length', hue='species', dodge=False)
plt.show()
```
