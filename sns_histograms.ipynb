```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# prevents warnings about future implementations from appearing
import warnings
warnings.filterwarnings('ignore', 'use_inf_as_na')

iris = sns.load_dataset('iris')
print(iris)
```

```python
ax = iris.plot.hist(column=['sepal_length'], bins=10, grid=True)
ax.set_title('Sepal Lengths')
ax.set_xlabel('Sepal length (cm)')
ax.set_ylabel('Iris count')
plt.show()
```

```python
print(iris['species'].unique())
print(iris.groupby('species').apply(lambda x:x))
```

```python
fix, ax = plt.subplots()
colors = {'setosa': 'Red', 'versicolor': 'Blue', 'virginica': 'Yellow'}
for species, data in iris.groupby('species'):
    # alpha .5 gives the columns transparency which helps see overlap
    data.plot(kind='hist', column='sepal_length', bins=15, alpha=.5, color=colors[species], ax=ax)
ax.set_xlabel('Sepal length (cm)')
ax.set_ylabel('Iris count')
ax.legend(colors.keys())
ax.set_title('Sepal Lengths by Iris species')
plt.show()
```

```python
# seaborn is built of matplotlib and can also generate graphs
sns.histplot(data=iris, x='sepal_length')
plt.show
```

```python
sns.histplot(data=iris, x='sepal_length', bins=25, hue='species')
plt.show
```

```python
# different 'elements' can be used to vary the chart
sns.histplot(data=iris, x='sepal_length', hue='species', element='step')
plt.show
```

```python
sns.histplot(data=iris, x='sepal_length', y='sepal_width', bins=20, hue='species')
plt.show
```

```python
# joint plot
graph = sns.JointGrid()
sns.histplot(data=iris, x='sepal_length', y='sepal_width', bins=20, hue='species', alpha=.6, ax=graph.ax_joint)
sns.histplot(data=iris, x='sepal_length', bins=20, hue='species', alpha=.6, ax=graph.ax_marg_x)
sns.histplot(data=iris, y='sepal_width', bins=20, hue='species', alpha=.6, ax=graph.ax_marg_y)
plt.show
```
