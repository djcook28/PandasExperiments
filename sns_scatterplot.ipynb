```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# prevents warnings about future implementations from appearing
import warnings
warnings.filterwarnings('ignore', 'use_inf_as_na')

iris = sns.load_dataset('iris')
print(iris)
```

```python
ax = iris.plot.scatter(x='petal_length', y='petal_width')
ax.set_xlabel='petal widths in cm'
ax.set_ylabel='petal lengths in cm'
ax.set_title('petal length v width scatter')
plt.show()
```

```python
fig, ax = plt.subplots()
colors = {'setosa': 'Red', 'versicolor': 'Purple', 'virginica': 'Yellow'}
for species, data in iris.groupby('species'):
    data.plot(kind='scatter', x='sepal_length', y='sepal_width', label=species, color=colors[species], ax=ax)
ax.set_xlabel='petal widths in cm'
ax.set_ylabel='petal lengths in cm'
ax.set_title('petal length v width scatter')
plt.show()
```

```python
fig, ax = plt.subplots()
colors = {'setosa': 'Red', 'versicolor': 'Purple', 'virginica': 'Yellow'}
for species, data in iris.groupby('species'):
    data.plot(kind='scatter', x='sepal_length', y='sepal_width', marker='+', label=species, color=colors[species], ax=ax)
    data.plot(kind='scatter', x='petal_length', y='petal_width', marker='x', label=species, color=colors[species], ax=ax)
ax.set_xlabel='petal widths in cm'
ax.set_ylabel='petal lengths in cm'
ax.set_title('petal length v width scatter')
plt.show()
```

```python
sns.scatterplot(data=iris, x='sepal_width', y='sepal_length', hue='species', style='species')
plt.show()
```

```python
sns.scatterplot(data=iris, x='sepal_width', y='sepal_length', hue='species', style='species', palette='bright')
plt.show()
```

```python
sns.pairplot(data=iris, hue='species')
plt.show()
```

```python
iris['sepal']='sepal'
iris['petal']='petal'
iris_1 = pd.DataFrame(data=iris[['sepal_length', 'sepal_width', 'species', 'sepal']])
iris_1.columns['length', 'width', 'species', 'measured']
iris_2 = pd.DataFrame(iris[['petal_length', 'petal_width', 'species', 'petal']])
iris_2.columns['length', 'width', 'species', 'measured']
iris2 = pd.concat([iris_1, iris_2)
sns.scatterplot(data=iris2, x='width', y='length', hue='species', style='measured')
plt.show()
```
